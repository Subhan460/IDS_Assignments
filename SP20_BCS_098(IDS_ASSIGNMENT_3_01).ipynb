```python
#Print the total number of students who have a three words name (first-middle-surname).
lsit = df['Name'].str.split(' ')
lsit
count = 0
for i in lsit:
    if len(i) >= 3:
        count = count+1
print(count)
```

```python
#Print the percentage of students who have a CGPA of 3.0 or above.
ad = 0
for i in df['CGPA']:
    j = float(i)
    if j >= 3.00:
        ad = ad + 1
per = (ad / 133) * 100
print(per)
```

```python
#Plot a pie chart to show the ratio of male and female students.
import matplotlib.pyplot as plt
import numpy as np
Male = 0
Female = 0


for i in df['Gender']:
    if i == "Male":
        Male = Male + 1
    else:
        Female = Female +1

#print(Male,Female)

Maleper  = (Male / 134)*100 
Femaleper = 100 - Maleper
#print(Maleper,Femaleper)
y = np.array([Maleper,Femaleper])

plt.pie(y)
plt.show() 
```
